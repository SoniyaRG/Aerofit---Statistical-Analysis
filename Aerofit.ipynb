```python
# Marginal Probability affecting Product based on Gender
# Calculate frequencies
joint_freq = pd.crosstab(df['Product'], df['Gender'])
product_freq = df['Product'].value_counts()
gender_freq = df['Gender'].value_counts()

# Calculate joint probabilities
joint_prob = joint_freq / joint_freq.sum().sum()

# Calculate marginal probabilities
product_prob = product_freq / product_freq.sum()
gender_prob = gender_freq / gender_freq.sum()

# Calculate conditional probabilities
# P(Gender|Product) = P(Gender âˆ© Product) / P(Product)
conditional_prob = joint_prob.div(product_prob, axis=0)


print("Conditional Probabilities P(Gender|Product):")
print(conditional_prob)
```

```python
# customer profilings for each and every product
products = df['Product'].unique()

# Dictionary to hold profiling results
profile_results = {}

# Compute profiles for each product
for product in products:
    df_product = df[df['Product'] == product]
    
    # Descriptive statistics
    age_stats = df_product['Age'].describe()
    income_stats = df_product['Income'].describe()
    gender_distribution = df_product['Gender'].value_counts(normalize=True)
    
    # Store results
    profile_results[product] = {
        'Age Statistics': age_stats,
        'Income Statistics': income_stats,
        'Gender Distribution': gender_distribution
    }


for product, profile in profile_results.items():
    print(f"\nProfile for Product {product}:")
    print("Age Statistics:")
    print(profile['Age Statistics'])
    print("\nIncome Statistics:")
    print(profile['Income Statistics'])
    print("\nGender Distribution:")
    print(profile['Gender Distribution'])
```
